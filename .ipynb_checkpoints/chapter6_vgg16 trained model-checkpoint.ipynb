### use trained model from network(model need to be downloaded)

```python
import tensorflow as tf
from tensorflow.python.keras.applications.vgg16 import VGG16
```

### model information

```python
model = VGG16()
model.summary()
```

### load image file  ***here need to "pip install pillow"  pillow module is necessary

```python
from tensorflow.python.keras.preprocessing.image import load_img
img_dog = load_img('img/dog1.jpg',target_size=(224,224))
img_cat = load_img('img/cat1.jpg',target_size=(224,224))
img_test = load_img('img/shrine.jpg',target_size=(224,224))
```

```python
img_dog
```

```python
img_cat
```

```python
img_test
```

### from image file to array data (pillow format to numpy.ndarray data format)

```python
from tensorflow.python.keras.preprocessing.image import img_to_array
arr_dog = img_to_array(img_dog)
arr_cat = img_to_array(img_cat)
arr_test = img_to_array(img_test)
```

```python
##arr_dog[0]
```

### 画像の各チャンネルの中心化とRGBからBGRへの変換

```python
from tensorflow.python.keras.applications.vgg16 import preprocess_input
arr_dog = preprocess_input(arr_dog)
arr_cat = preprocess_input(arr_cat)
arr_test = preprocess_input(arr_test)
##arr_dog[0]
```

### 2枚の画像を一つのデータにまとめる

```python
import numpy as np
arr_input = np.stack([arr_dog,arr_cat,arr_test])
print('shape of arr_input:', arr_input.shape)
```

### data input to vgg16 model and predict the result. vgg16 output is 1000 labels, so the total output is 2*1000

```python
probs = model.predict(arr_input)
print('shape of probs:', probs.shape)
```

```python
##list(probs[0])
```

```python
list(probs[0]).index(max(probs[0]))
```

### 通过输出的概率来反推具体的标签, 需要下载标签数据和标签内容对应文件,此函数仅输出top5位

```python
from tensorflow.python.keras.applications.vgg16 import decode_predictions
results = decode_predictions(probs)
```
